#Popularity Based Recommender System

```python
num_of_rating = rating_book.groupby('Book-Title').count()['Book-Rating'].reset_index()
num_of_rating.rename(columns={'Book-Rating':'num_rating'},inplace=True)
num_of_rating
```

```python
avg_rating = rating_book.groupby('Book-Title')['Book-Rating'].mean().round(2).reset_index()
avg_rating
```

##Merge datasets

```python
popular_data = num_of_rating.merge(avg_rating, on='Book-Title')
popular_data
```

```python
popular_data = popular_data[popular_data['num_rating'] >= 150].sort_values('Book-Rating', ascending=False).head(50)
popular_data.head()
```

```python
popular_data = popular_data.merge(books , on='Book-Title').drop_duplicates('Book-Title')[['Book-Title','Book-Author','Image-URL-M','num_rating','Book-Rating']]
```

```python
popular_data.shape
```

```python
popular_data
```

#Preprocess data

```python
data1 = rating_book.groupby('User-ID').count()['Book-Rating'] > 200
exp_users = data1[data1].index
```

```python
filtered_rating = rating_book[rating_book['User-ID'].isin(exp_users)]
```

```python
data2 = filtered_rating.groupby('Book-Title').count()['Book-Rating'] >= 50
famous_books = data2[data2].index
famous_books
```

```python
final_ratings = filtered_rating[filtered_rating['Book-Title'].isin(famous_books)]
```

```python
pt = final_ratings.pivot_table(index='Book-Title', columns='User-ID', values='Book-Rating')
pt.fillna(0, inplace=True)
```

```python
pt
```

```python
pt.loc["Violets Are Blue"][:5]
```

#KNN

```python
from sklearn.neighbors import NearestNeighbors
```

```python
model = NearestNeighbors(metric='cosine')
model.fit(pt.values)
```

```python
title = 'Violets Are Blue'
pt.loc[title].shape
```

```python
distance, indice = model.kneighbors([pt.loc[title].values], n_neighbors=5)
print(distance)
print(indice)
```

```python
recommend = pt.iloc[indice[0]].index.values
```

```python
list(recommend)
```

```python

def recommend(book_name, k=5):

    index = np.where(pt.index == book_name)[0][0]


    distances, indices = model.kneighbors(pt.iloc[index, :].values.reshape(1, -1), n_neighbors=k+1)

    data = []
    for i in range(0, len(distances.flatten())):
        item = []
        temp_df = books[books['Book-Title'] == pt.index[indices.flatten()[i]]]

        item.extend(list(temp_df.drop_duplicates('Book-Title')['Book-Title'].values))
        item.extend(list(temp_df.drop_duplicates('Book-Title')['Book-Author'].values))
        item.extend(list(temp_df.drop_duplicates('Book-Title')['Image-URL-M'].values))

        data.append(item)

    return data
```

```python
recommend('Violets Are Blue')
```
